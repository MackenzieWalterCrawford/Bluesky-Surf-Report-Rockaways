```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://surfcaptain.com/forecast/rockaway-new-york'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
summary = soup.find('h1', id="fcst-current-title")
```

```python
full_summary = ""
for ele in summary.contents:
    print(type(ele))
```

```python
full_summary = []
for ele in summary.contents:
    print(ele.text.strip())
    full_summary.append(ele.text.strip())
```

```python
surf_output = " ".join(full_summary)
```
